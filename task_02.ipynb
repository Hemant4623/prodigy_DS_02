Necessary Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Load the Iris dataset

```python
iris = sns.load_dataset('iris')
iris.head()
```

check for missing values

```python
iris.isnull().sum()
```

```python
# check data type
iris.info()
```

Summary Statistics

```python
iris.describe()
```

Visualize Pairwise Relationships

```python
sns.pairplot(iris, hue='species')
plt.show()
```

Correlation Matrix


```python
# Select only the numeric columns for correlation calculation
numeric_iris = iris.select_dtypes(include=['float', 'int'])

plt.figure(figsize=(10, 8))
sns.heatmap(numeric_iris.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

Box Plot

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=iris)
plt.title('Box Plot of Iris Features')
plt.show()
```

Violin Plot

```python
plt.figure(figsize=(12, 6))
sns.violinplot(x='species', y='sepal_length', data=iris)
plt.title('Violin Plot of Sepal Length by Species')
plt.show()
```

Scatter Plot with Linear Regression

```python
sns.lmplot(x='sepal_length', y='sepal_width', data=iris, hue='species', markers=['o', 's', 'D'])
plt.title('Sepal Length vs Sepal Width with Linear Regression')
plt.show()
```
